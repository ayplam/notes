```python
# Dataframe tools
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import math
from datetime import datetime
import pickle
import random
# MPL and D3
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

```python
class Perceptron:
    
    def __init__(self, nwgts):
        
        self.wgts = []
        # Learning constant (determins how easily weights can be changed)
        self.c = 0.02
        
        for cnt in xrange(nwgts):
            # Initialize random weights
            self.wgts.append(np.sign(random.uniform(-1,1)))
            

    def feedforward(self,inputs):
        # take the dot product of the inputs and weights. The sign determines the output
        return np.sign(sum([inp * wgt for (inp,wgt) in zip(inputs,self.wgts)]))

    def train(self,inputs, actual):
        
        pred = self.feedforward(inputs)
        err = actual-pred

        for x in xrange(len(inputs)):
            # Iteratively update the weights
            self.wgts[x] += self.c * err * inputs[x]
            
class Trainer:
    
    def __init__(self,x,y,actual):
        
        self.inputs = []
        self.inputs.append(x)
        self.inputs.append(y)
        self.inputs.append(-1) # Bias in case values are zero
        self.actual = actual

def createData():

    x = random.uniform(0,5)
    y = random.uniform(0,5)

    if line(x) > y:
        ans = 1
    else:
        ans = -1
        
    return x,y,ans
        
    
def line(x):
    return 2*x + 3

def createLine(wgts):
    # Line will be created in form y = mx + b
    x = np.linspace(0,5,20)
    # multiple the second weight by the bias
    y = x * -wgts[0]/wgts[1] + wgts[2] * (-1) 
    return x,y
```

```python
ptron = Perceptron(3)
cmap = sns.color_palette("RdBu_r", 20)
print len(cmap)


for x in xrange(1000000):
    x1,y1,ans = createData()
    t = Trainer(x1,y1,ans)
    
    ptron.train(t.inputs,t.actual)
    
    if (x < 10000) and (x%500 == 0):
        x2,y2 = createLine(ptron.wgts)
        plt.plot(x2,y2,'-',color=cmap[x/500])
        
plt.plot(np.linspace(0,5,20),2*np.linspace(0,5,20) + 1,'g-')

plt.gca().set_ylim([-5,5])
```
